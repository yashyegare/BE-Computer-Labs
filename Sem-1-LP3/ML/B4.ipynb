# Importing Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Gradient Descent

```python
# Define the function y = (x+3)^2
def function(x):
 return (x + 3) ** 2

# Define the derivative of the function
def derivative(x):
 return 2 * (x + 3)

# Gradient Descent Algorithm
def gradient_descent(initial_x, learning_rate, iterations):
    x_values = [initial_x]
    y_values = [function(initial_x)]
    
    x = initial_x
    for i in range(iterations):
        grad = derivative(x)
        x = x - learning_rate * grad
        x_values.append(x)
        y_values.append(function(x))
        
        # Print values for each iteration
        print(f"Iteration {i+1}: x = {x}, y = {function(x)}")
    return x_values, y_values

# Initial parameters
initial_x = 2 # Initial point
learning_rate = 0.1 # Learning rate
iterations = 15 # Number of iterations

# Run gradient descent
x_vals, y_vals = gradient_descent(initial_x, learning_rate, iterations)

# Plotting the function
x = np.linspace(-10, 4, 1000) # Adjusted the range to better fit the funct
y = function(x)
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='y = (x+3)^2')
plt.scatter(x_vals, y_vals, color='red', label='Gradient Descent Steps')
plt.title('Gradient Descent for Finding Local Minimum')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
    
```
